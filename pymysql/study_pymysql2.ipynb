```python
from db_util import get_db
db = get_db()
```

```python
# 3. 커서 가져오기
cursor = db.cursor()
```

```python
sql = '''
CREATE TABLE IF NOT EXISTS product2 (
PRODUCT_CODE VARCHAR(20) NOT NULL,
TITLE VARCHAR(200) NOT NULL,
ORI_PRICE INT,
DISCOUNT_PRICE INT,
DISCOUNT_PERCENT INT,
DELIVERY VARCHAR(2),
PRIMARY KEY(PRODUCT_CODE)
);
'''
```

```python
# 5. SQL 구문 실행하기
cursor.execute(sql)
```
