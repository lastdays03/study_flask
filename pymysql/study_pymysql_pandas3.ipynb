```python
from db_util import get_db
import pandas as pd

db = get_db()
cursor = db.cursor()
```

```python
# 여러 쿼리문이 들어있는 sqlDB.sql 파일을 읽기
with open('sqlDB.sql', 'r', encoding='utf-8') as f:
    sql_file = f.read()
    sql_querys = sql_file.split(";")
```

```python
# 각각의 쿼리문 실행
for query in sql_querys:
    query = query.strip()
    if query:
        cursor.execute(query)
```
